# QUESITO 3

### Definisco la funzione del quesito 3:

```python
def confronto_casi_totali_tra_continenti(df_covid, continente1, continente2):
    df_3_contienti = df_covid[df_covid['continent'].notnull()].groupby(['continent'])['new_cases'].sum().reset_index()
    df_3_contienti = df_3_contienti.rename(columns ={'new_cases':'computed_total_cases'})
    
    
    casi_totali_continente_1 = int(df_3_contienti[df_3_contienti['continent']==continente1]['computed_total_cases'])
    casi_totali_continente_2 = int(df_3_contienti[df_3_contienti['continent']==continente2]['computed_total_cases'])
    casi_totali_world = int(df_covid[df_covid['location']=='World']['new_cases'].sum())
    
    confronto_val_min = min(casi_totali_continente_1,casi_totali_continente_2)
    confronto_val_max = max(casi_totali_continente_1,casi_totali_continente_2)
    confronto_val_avg = (casi_totali_continente_1+casi_totali_continente_2)/2
    perc_1_vs_world = round(100*(casi_totali_continente_1/casi_totali_world),4)
    perc_2_vs_world = round(100*(casi_totali_continente_2/casi_totali_world),4)
    
    print('Numero casi totali %s: %d'%(continente1,casi_totali_continente_1))
    print('Numero casi totali %s: %d'%(continente2,casi_totali_continente_2))
    print('Numero casi totali mondo: %d'%(casi_totali_world))
    
    print('\n')
    print('Valore minimo: %d'%(confronto_val_min))
    print('Valore massimo: %d'%(confronto_val_max))
    print('Valore medio: %d'%(confronto_val_avg))    
    print('\n')
    print('percentuale casi %s vs World: %f'%(continente1,perc_1_vs_world)+' %')
    print('percentuale casi %s vs World: %f'%(continente2,perc_2_vs_world)+' %')
    
```

### Esempio:

```python
#esempio: Asia vs Oceania
confronto_casi_totali_tra_continenti(df_covid,'Asia','Oceania')
```

```python
def confronto_vaccinazioni_totali_tra_continenti(df_covid, continente1, continente2,continente3, continente4,continente5,continente6):
    df_3_contienti = df_covid[df_covid['continent'].notnull()].groupby(['continent'])['new_vaccinations'].sum().reset_index()
    df_3_contienti = df_3_contienti.rename(columns ={'new_vaccinations':'computed_total_vaccinations'})
    
    
    vaccinazioni_totali_continente_1 = int(df_3_contienti[df_3_contienti['continent']==continente1]['computed_total_vaccinations'])
    vaccinazioni_totali_continente_2 = int(df_3_contienti[df_3_contienti['continent']==continente2]['computed_total_vaccinations'])
    vaccinazioni_totali_continente_3 = int(df_3_contienti[df_3_contienti['continent']==continente3]['computed_total_vaccinations'])
    vaccinazioni_totali_continente_4 = int(df_3_contienti[df_3_contienti['continent']==continente4]['computed_total_vaccinations'])
    vaccinazioni_totali_continente_5 = int(df_3_contienti[df_3_contienti['continent']==continente5]['computed_total_vaccinations'])
    vaccinazioni_totali_continente_6 = int(df_3_contienti[df_3_contienti['continent']==continente6]['computed_total_vaccinations'])
    vaccinazioni_totali_world = int(df_covid[df_covid['location']=='World']['new_vaccinations'].sum())
    
    confronto_val_min = min(vaccinazioni_totali_continente_1,vaccinazioni_totali_continente_2,vaccinazioni_totali_continente_3, vaccinazioni_totali_continente_4,vaccinazioni_totali_continente_5,vaccinazioni_totali_continente_6)
    confronto_val_max = max(vaccinazioni_totali_continente_1,vaccinazioni_totali_continente_2,vaccinazioni_totali_continente_3, vaccinazioni_totali_continente_4,vaccinazioni_totali_continente_5,vaccinazioni_totali_continente_6)
    confronto_val_avg = (vaccinazioni_totali_continente_1+vaccinazioni_totali_continente_2+vaccinazioni_totali_continente_3+ vaccinazioni_totali_continente_4+vaccinazioni_totali_continente_5+vaccinazioni_totali_continente_6)/5
    perc_1_vs_world = round(100*(vaccinazioni_totali_continente_1/vaccinazioni_totali_world),4)
    perc_2_vs_world = round(100*(vaccinazioni_totali_continente_2/vaccinazioni_totali_world),4)
    perc_3_vs_world = round(100*(vaccinazioni_totali_continente_3/vaccinazioni_totali_world),4)
    perc_4_vs_world = round(100*(vaccinazioni_totali_continente_4/vaccinazioni_totali_world),4)
    perc_5_vs_world = round(100*(vaccinazioni_totali_continente_5/vaccinazioni_totali_world),4)
    perc_6_vs_world = round(100*(vaccinazioni_totali_continente_6/vaccinazioni_totali_world),4)
    
    print('Numero vaccinazioni totali %s: %d'%(continente1,vaccinazioni_totali_continente_1))
    print('Numero vaccinazioni totali %s: %d'%(continente2,vaccinazioni_totali_continente_2))
    print('Numero vaccinazioni totali %s: %d'%(continente3,vaccinazioni_totali_continente_3))
    print('Numero vaccinazioni totali %s: %d'%(continente4,vaccinazioni_totali_continente_4))
    print('Numero vaccinazioni totali %s: %d'%(continente5,vaccinazioni_totali_continente_5))
    print('Numero vaccinazioni totali %s: %d'%(continente6,vaccinazioni_totali_continente_6))
    
    print('Numero vaccinazioni totali mondo: %d'%(vaccinazioni_totali_world))
    
    print('\n')
    print('Valore minimo: %d'%(confronto_val_min))
    print('Valore massimo: %d'%(confronto_val_max))
    print('Valore medio: %d'%(confronto_val_avg))    
    print('\n')
    print('percentuale vaccinazioni %s vs World: %f'%(continente1,perc_1_vs_world)+' %')
    print('percentuale vaccinazioni %s vs World: %f'%(continente2,perc_2_vs_world)+' %')
    print('percentuale vaccinazioni %s vs World: %f'%(continente3,perc_3_vs_world)+' %')
    print('percentuale vaccinazioni %s vs World: %f'%(continente4,perc_4_vs_world)+' %')
    print('percentuale vaccinazioni %s vs World: %f'%(continente5,perc_5_vs_world)+' %')   
    print('percentuale vaccinazioni %s vs World: %f'%(continente6,perc_6_vs_world)+' %')   
    
```

```python
def confronto_casi_totali_tra_continenti_per_confronto(df_covid, continente1, continente2,continente3):
    df_3_contienti = df_covid[df_covid['continent'].notnull()].groupby(['continent'])['new_cases'].sum().reset_index()
    df_3_contienti = df_3_contienti.rename(columns ={'new_cases':'computed_total_cases'})
    
    
    casi_totali_continente_1 = int(df_3_contienti[df_3_contienti['continent']==continente1]['computed_total_cases'])
    casi_totali_continente_2 = int(df_3_contienti[df_3_contienti['continent']==continente2]['computed_total_cases'])
    casi_totali_continente_3 = int(df_3_contienti[df_3_contienti['continent']==continente3]['computed_total_cases'])
    casi_totali_world = int(df_covid[df_covid['location']=='World']['new_cases'].sum())
    
    confronto_val_min = min(casi_totali_continente_1,casi_totali_continente_2,casi_totali_continente_3)
    confronto_val_max = max(casi_totali_continente_1,casi_totali_continente_2,casi_totali_continente_3)
    confronto_val_avg = (casi_totali_continente_1+casi_totali_continente_2+casi_totali_continente_3)/3
    perc_1_vs_world = round(100*(casi_totali_continente_1/casi_totali_world),4)
    perc_2_vs_world = round(100*(casi_totali_continente_2/casi_totali_world),4)
    perc_3_vs_world = round(100*(casi_totali_continente_3/casi_totali_world),4)
    
    print('Numero casi totali %s: %d'%(continente1,casi_totali_continente_1))
    print('Numero casi totali %s: %d'%(continente2,casi_totali_continente_2))
    print('Numero casi totali %s: %d'%(continente3,casi_totali_continente_3))
    print('Numero casi totali mondo: %d'%(casi_totali_world))
    
    print('\n')
    print('Valore minimo: %d'%(confronto_val_min))
    print('Valore massimo: %d'%(confronto_val_max))
    print('Valore medio: %d'%(confronto_val_avg))    
    print('\n')
    print('percentuale casi %s vs World: %f'%(continente1,perc_1_vs_world)+' %')
    print('percentuale casi %s vs World: %f'%(continente2,perc_2_vs_world)+' %')
    print('percentuale casi %s vs World: %f'%(continente3,perc_3_vs_world)+' %')
    
```
